```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mall_data_path= "/content/drive/MyDrive/Mall_Customers.csv"
df=pd.read_csv(mall_data_path)
df.head()
```

```python
data=df.drop("CustomerID",axis=1)
data["Genre"]=data["Genre"].map({"Male":0,"Female":1})
data.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
X_scaled=scaler.fit_transform(data)
X_scaled[:5]
```

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
```

```python
dbscan=DBSCAN(eps=0.8,min_samples=5)
labels=dbscan.fit_predict(X_scaled)
```

```python
df["DBSCAN_Cluster"] = labels
df.head()
```

```python
print(df["DBSCAN_Cluster"].value_counts())
num_clusters=len(set(labels))-(1 if -1 in labels else 0)
num_noise=list(labels).count(-1)
```

```python
print("Number of clusters:",num_clusters)
print("Number of noise points:",num_noise)
```

```python
if num_clusters>1:
    mask=labels!= -1
    score=silhouette_score(X_scaled[mask],labels[mask])
    print("Silhouette Score without noise:",score)
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=df,
    x="Annual Income (k$)",
    y="Spending Score (1-100)",
    hue="DBSCAN_Cluster",
    palette="Set2",
    s=80
)

plt.title("DBSCAN Clustering on Mall Customers")
plt.xlabel("Annual Income(k$)")
plt.ylabel("Spending Score (1-100)")
plt.legend(title="Cluster")
plt.show()
```

```python
from sklearn.decomposition import PCA

pca=PCA(n_components=2)
X_pca=pca.fit_transform(X_scaled)

pca_df=pd.DataFrame(X_pca,columns=["PC1","PC2"])
dbscan_pca=DBSCAN(eps=0.8, min_samples=5)
labels_pca=dbscan_pca.fit_predict(X_pca)
pca_df["DBSCAN_Cluster"]=labels_pca
pca_df.head()
```

```python
total_variance = pca.explained_variance_ratio_.sum()
total_variance
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=pca_df,
    x="PC1",
    y="PC2",
    hue="DBSCAN_Cluster",
    palette="Set2",
    s=80
)

plt.title("DBSCAN Clusters After PCA")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend(title="Cluster")
plt.show()
```

```python
total_variance = pca.explained_variance_ratio_.sum()
total_variance
```

```python
num_clusters_pca=len(set(labels_pca))-(1 if -1 in labels_pca else 0)
num_noise_pca=list(labels_pca).count(-1)

print("Clusters (PCA):",num_clusters_pca)
print("Noise (PCA):",num_noise_pca)
```

```python
print("Original clusters:", num_clusters)
print("Original noise:", num_noise)

print("PCA clusters:", num_clusters_pca)
print("PCA noise:", num_noise_pca)
```
